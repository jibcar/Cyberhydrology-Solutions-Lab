```python
prec_max_data=pd.read_csv("C:/Users/joseph57/Box/01_SharedBox_Jibin/Results/prec_noaa_analysis/input_data2/A_Result_prec_prec_results_max_new.txt",
                     delimiter='\t')
prec_max_data["Longitude2"] = prec_max_data["Longitude"]-360
prec_max_data["Norm_MK_Combi_slope"] = prec_max_data['MK_Combi_slope']/prec_max_data['StdDev']


# Create MultiIndex using both latitude and longitude
prec_max_data.set_index(['Latitude', 'Longitude2'], inplace=True)
print(prec_max_data.columns)
#print(prec_max_data.head())

list_grid_points=prec_max_data.index.tolist()
print(len(list_grid_points))
print((prec_max_data.shape))
```

```python
prec_mean_data=pd.read_csv("C:/Users/joseph57/Box/01_SharedBox_Jibin/Results/prec_noaa_analysis/input_data2/A_Result_prec_prec_results_mean_new.txt",
                     delimiter='\t')
prec_mean_data["Longitude2"] = prec_mean_data["Longitude"]-360
prec_mean_data["Norm_MK_Combi_slope"] = prec_mean_data['MK_Combi_slope']/prec_mean_data['StdDev']


# Create MultiIndex using both latitude and longitude
prec_mean_data.set_index(['Latitude', 'Longitude2'], inplace=True)
print(prec_mean_data.head())
```

```python
res=0.25

#watersheds["Weigh_Max_TS"]=""
#watersheds["Weigh_Mean_TS"]=""

ii=1

# Iterate over each watershed
for index, watershed in watersheds.iterrows():
    print(f"{ii}",end= '-')
    print(f"{watershed['USGS_St']}",end= '; ')
    
    # Get the bounding box (extents) of the watershed
    minx, miny, maxx, maxy = watershed.geometry.bounds
        
    # Print the extents of the watershed
    #print(f"Watershed {index+1}: {watershed['USGS_St']}")
    #print(f"Minimum X-coordinate: {minx}")
    #print(f"Minimum Y-coordinate: {miny}")
    #print(f"Maximum X-coordinate: {maxx}")
    #print(f"Maximum Y-coordinate: {maxy}")
    
    #print(x for x in specified_x_values if x >= maxx)
    
    # Ceil and floor the minimum x and y coordinates to the specified values
    min_lon_value = max(value for value in specified_x_values if value <= minx)
    #print("Min Lat",min_lon_value)
    max_lon_value = min(value for value in specified_x_values if value >= maxx)
    #print("Max Lat",max_lon_value)
    min_lat_value = max(value for value in specified_y_values if value <= miny)
    #print("Min Lat",min_lat_value)
    max_lat_value = min(value for value in specified_y_values if value >= maxy)
    #print("Max Lat",max_lat_value)
    

    polygons = []
    points=[]
    lon_list=np.arange(min_lon_value,max_lon_value+res,res).tolist()
    lat_list=np.arange(min_lat_value,max_lat_value+res,res).tolist()
    for lon in lon_list:#range(min_lon_value,max_lon_value,0.25):
        for lat in lat_list:#range(min_lat_value,max_lat_value,0.25):
            if (lat,lon) in list_grid_points:
                #print("Yes")
            
            
                current_lon = lon
        #while current_lon < max_lon_value:
                current_lat = lat
            #while current_lat < max_lat_value:
                #print("Current:",current_lat,current_lon)
                #print("Max Prec Slope",prec_max_data.loc[(lat,lon),'Ori_Trend_slope'])
                #print("Min Prec Slope",prec_mean_data.loc[(lat,lon),'Ori_Trend_slope'])
                polygon = Polygon([(current_lon-res/2, current_lat-res/2), (current_lon+res/2,current_lat-res/2),\
                                   (current_lon+res/2, current_lat+res/2), (current_lon-res/2, current_lat+res/2)])
                point=Point(lon,lat)
                # Add latitude and longitude fields
                lat_centroid, lon_centroid = polygon.centroid.y, polygon.centroid.x
                #polygon_data = {'geometry': polygon, 'lat': lat, 'lon': lon}
                polygons.append({'geometry': polygon, 'lat': current_lat, 'lon': current_lon,\
                                 'Max_Trend_Slope': prec_max_data.loc[(lat,lon),'MK_Combi_slope'],\
                                 'Mean_Trend_Slope': prec_mean_data.loc[(lat,lon),'MK_Combi_slope'],
                                 'Norm_Max_Trend_Slope': prec_max_data.loc[(lat,lon),'Norm_MK_Combi_slope'],\
                                 'Norm_Mean_Trend_Slope': prec_mean_data.loc[(lat,lon),'Norm_MK_Combi_slope'],
                                 'Max_Trend_Result': prec_max_data.loc[(lat,lon),'MK_Combi_Trend'],
                                 'Mean_Trend_Result': prec_mean_data.loc[(lat,lon),'MK_Combi_Trend'],
                                 'Max_Std_Dev': prec_max_data.loc[(lat,lon),'StdDev'],
                                 'Mean_Std_Dev': prec_mean_data.loc[(lat,lon),'StdDev'],
                                 'Grid_Mean_Prec': grid_mean_data.loc[(lat,lon),'precip']
                                })
                points.append({'geometry': point, 
                               'lat': current_lat, 
                               'lon': current_lon})
            else:
                continue

    polygon_gdf=gpd.GeoDataFrame(polygons,crs="EPSG:4326")
    point_gdf=gpd.GeoDataFrame(points,crs="EPSG:4326")
    #print("Done1")

    # Specify the extent (xmin, ymin, xmax, ymax)
    #extent = (-extent_left,extent_bottom, -extent_right, extent_top)

    # Generate random polygons within the extent
    #gdf = generate_polygons(extent, 100)

    # Save the generated polygons to a shapefile
    #print({watershed['USGS_St']})
    #gdf.to_file(f"generated_{watershed['USGS_St']}.shp")
    
    single_watershed_gdf = gpd.GeoDataFrame([watershed], crs=watersheds.crs)
    
    # Perform the intersection
    intersection = gpd.overlay(polygon_gdf, single_watershed_gdf, how='intersection')
    
    # Calculate the area of each polygon
    intersection['Area'] = intersection['geometry'].area
    
    


    """
    # Plot both shapefiles
    fig, ax = plt.subplots(1,2,figsize=(12,6))
    
    polygon_gdf.plot(ax=ax[0], color='grey', alpha=0.5, edgecolor='black', label='Generated Polygons')
    point_gdf.plot(ax=ax[0], color='green')
    single_watershed_gdf.plot(ax=ax[0], color='red', alpha=0.5, edgecolor='black', label='Other Shapefile')
    
    ax[0].set_title(f"Generated Polygons, NOAA Grid, and Watershed File\n(for {watershed['USGS_St']})")
    ax[0].set_xlabel('Longitude')
    ax[0].set_ylabel('Latitude')
    # Set tick locations and labels for the first subplot
    ax[0].set_xticks(lon_list)
    ax[0].set_xticklabels(lon_list, rotation=0)
    ax[0].set_yticks(lat_list)
    ax[0].set_yticklabels(lat_list, rotation=0)
    
    polygon_gdf.plot(ax=ax[1], color='grey', alpha=0.5, edgecolor='black', label='Generated Polygons')
    point_gdf.plot(ax=ax[1], color='green')
    intersection.plot(ax=ax[1],column="Area")  # Plot the intersection
    ax[1].set_title(f"Generated Polygons, NOAA Grid, and Intersected Areas\n(for {watershed['USGS_St']})")
    ax[1].set_xlabel('Longitude')
    ax[1].set_ylabel('Latitude')
    ax[1].set_xticks(lon_list)
    ax[1].set_xticklabels(lon_list, rotation=0)
    ax[1].set_yticks(lat_list)
    ax[1].set_yticklabels(lat_list, rotation=0)
    
    plt.show()
    """
    #print(intersection)
    intersection['Area x Max_Trend_Slope'] = intersection['Max_Trend_Slope']*intersection['Area']
    intersection['Area x Mean_Trend_Slope'] = intersection['Mean_Trend_Slope']*intersection['Area']
    
    intersection['Area x Norm_Max_Trend_Slope'] = intersection['Norm_Max_Trend_Slope']*intersection['Area']
    intersection['Area x Norm_Mean_Trend_Slope'] = intersection['Norm_Mean_Trend_Slope']*intersection['Area']
    
    intersection['Area x Max_Std_Dev'] = intersection['Max_Std_Dev']*intersection['Area']
    intersection['Area x Mean_Std_Dev'] = intersection['Mean_Std_Dev']*intersection['Area']
    
    intersection['Area x Grid_Mean_Prec'] = intersection['Grid_Mean_Prec']*intersection['Area']

    
    max_weigh_value=intersection['Area x Max_Trend_Slope'].sum()/intersection['Area'].sum()
    mean_weigh_value=intersection['Area x Mean_Trend_Slope'].sum()/intersection['Area'].sum()
    
    max_weigh_value2=intersection['Area x Norm_Max_Trend_Slope'].sum()/intersection['Area'].sum()
    mean_weigh_value2=intersection['Area x Norm_Mean_Trend_Slope'].sum()/intersection['Area'].sum()
    
    max_weigh_value3=intersection['Area x Max_Std_Dev'].sum()/intersection['Area'].sum()
    mean_weigh_value3=intersection['Area x Mean_Std_Dev'].sum()/intersection['Area'].sum()
    
    grid_mean_weigh_value=intersection['Area x Grid_Mean_Prec'].sum()/intersection['Area'].sum()
    
    #print("Weighted Max Trend Slope for watershed:\t",max_weigh_value)
    #print("Weighted Mean Trend Slope for watershed:\t",mean_weigh_value)
    
    
    
    # Update the GeoDataFrame with the multiplied area
    #gdf.at[idx, 'area_multiplied'] = area_multiplied
    
    ## Initialize counts for "Max Prec"
    count_inc_max = 0
    count_dec_max = 0
    count_not_max = 0

    # Iterate through the values in 'col1' and count occurrences of "A" and "B"
    for value in intersection['Max_Trend_Result']:
        if value == "increasing":
            count_inc_max += 1
        elif value == "decreasing":
            count_dec_max += 1
        elif value == "no trend":
            count_not_max += 1

    #print("Number of increasings:", count_inc_max)
    #print("Number of decreasings:", count_dec_max)
    #print("Number of no_trends:", count_not_max)
    
    ## Initialize counts for "Max Prec"
    count_inc_mean = 0
    count_dec_mean = 0
    count_not_mean = 0

    # Iterate through the values in 'col1' and count occurrences of "A" and "B"
    for value in intersection['Mean_Trend_Result']:
        if value == "increasing":
            count_inc_mean += 1
        elif value == "decreasing":
            count_dec_mean += 1
        elif value == "no trend":
            count_not_mean += 1

    #print("Number of increasings:", count_inc_mean)
    #print("Number of decreasings:", count_dec_mean)
    #print("Number of no_trends:", count_not_mean)

    watersheds.at[index,"Weigh_Max_TS"]=max_weigh_value
    watersheds.at[index,"Weigh_Mean_TS"]=mean_weigh_value
    watersheds.at[index,"Weigh_Norm_Max_TS"]=max_weigh_value2
    watersheds.at[index,"Weigh_Norm_Mean_TS"]=mean_weigh_value2
    watersheds.at[index,"Weigh_Max_SD"]=max_weigh_value3
    watersheds.at[index,"Weigh_Mean_SD"]=mean_weigh_value3
    watersheds.at[index,"percent_max_increase"]=count_inc_max/(count_inc_max+count_dec_max+count_not_max)*100
    watersheds.at[index,"percent_max_decrease"]=count_dec_max/(count_inc_max+count_dec_max+count_not_max)*100
    watersheds.at[index,"percent_max_notrend"]=count_not_max/(count_inc_max+count_dec_max+count_not_max)*100
    watersheds.at[index,"percent_mean_increase"]=count_inc_mean/(count_inc_mean+count_dec_mean+count_not_mean)*100
    watersheds.at[index,"percent_mean_decrease"]=count_dec_mean/(count_inc_mean+count_dec_mean+count_not_mean)*100
    watersheds.at[index,"percent_mean_notrend"]=count_not_mean/(count_inc_mean+count_dec_mean+count_not_mean)*100
    
    watersheds.at[index,"ws_mean_precip"]=grid_mean_weigh_value
    
    
    # Save or plot the result as desired
    #intersection.to_file(f"intersection_{watershed['USGS_St']}.shp")  # Save to a shapefile
    
    #print(intersection)
    #print(intersection.columns)
    
    
    #print("###########################################################\n\n")
    
    #print("Done2")
    ii+=1
    
#watersheds.to_file("watersheds5_3890.shp")
print("Completed.........")
```
